# Pine script EMA

Under er et eksempel på et script hvor en klient ønsket et script som ga alarm nå to EMA-er med forskjellig tidsintervaller krysset hverandre. Pine script er utviklet av og for Tradingveiw. Tradingveiw er en interaktiv trading plattform 

// This source code is subject to the terms of the Mozilla Public License 2.0 at https://mozilla.org/MPL/2.0/
//@version=4

study("EMA with put/call alarms", overlay=true,max_bars_back=1000)




fast = input(10, minval=1, title="fast ema")
fast_ema = ema(close, fast)

slow = input(50, minval=1, title="slow ema")
slow_ema = ema(close, slow)

over = crossover(fast_ema,slow_ema)
under = crossunder(fast_ema,slow_ema)


if (over)
    lbl = label.new(bar_index,low,"Call")
    label.set_color(lbl,color.green)
    label.set_yloc(lbl,yloc.belowbar)
    label.set_style(lbl,label.style_label_up)


else if (under)
    lbl = label.new(bar_index,low,"Put")
    label.set_color(lbl,color.red)
    label.set_yloc(lbl,yloc.belowbar)
    label.set_style(lbl,label.style_label_up)



plot(fast_ema,color=#ff4000)
plot(slow_ema)


Hvis du har en Tradingveiw konto kan du selv kjøre koden over for å se resultatet, eller du kan se et utklipp av det under.
